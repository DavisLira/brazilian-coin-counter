```python
import matplotlib.pyplot as plt
from skimage.io import imread

image_path = '../images/coin_01.jpg' 

try:
    img = imread(image_path)
    plt.imshow(img)
    plt.title("Imagem Carregada com Sucesso")
    plt.axis('off')
    plt.show()
except FileNotFoundError:
    print(f"Erro: No foi encontrada a imagem em {image_path}. Verifique o caminho!")
```
